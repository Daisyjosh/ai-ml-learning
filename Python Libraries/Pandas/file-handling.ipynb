```python
import pandas as pd
import numpy as np
```

```python
# CSV data

df = pd.read_csv("raw_data.csv")
print(df)
print(type(df))
```

```python
# json data

df = pd.read_json("employee_data.json")
print(df)
print(type(df))
```

```python
# DataFrame Methods

# head() - prints starting 5 rows
print("prints starting 5 rows") # default five rows
print(df.head())
print("prints starting n rows")
print(df.head(2))

# tail() - trailing 5 rows
print("prints trailing 5 rows") # default 5 rows
print(df.tail())
print("prints trailing n rows")
print(df.tail(3))

# sample() - random row
print("prints a sample row") # default 1 row
print(df.sample())
print(df.sample(2))
```

```python
# info() - summary of the data
print(df.info())

# shape - tuple of (rows,columns)
print("Printing number of rows and columns from the given data")
print(df.shape)

# describe() - mathematical description of the data
print(df.describe())

# columns() - columns in the data
print(df.columns)

# unique data
print(df.nunique())
```
