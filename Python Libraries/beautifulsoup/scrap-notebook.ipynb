```python
# BeautifulSoup(bs4) is a python library used to prase HTML and XML documents.
# It helps you extract data from webpages after downloading them with requests.
# It creates a Parse tree from the data from which we can extract text. attributes, links, tables etc.
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
# scrapping the data
URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL)
print(res)
```

```python
# saving the scrapped data 
with open("scraped-data/data1.html","w",encoding="utf-8") as f:
    f.write(res.text)
```

```python
# parse the data
with open("scraped-data/data1.html","r") as f:
    html_content = f.read()
soup = BeautifulSoup(html_content,"lxml") 
```

```python
# parse the  required data from the html page and use it for ML model

# to find the first occurence of a tag
soup.find("h1")
soup.h1 # works the same way
```

```python
soup.find("h3")
soup.h3 # works the same way
```

```python
# to find all the occurence
all_h3 = soup.find_all("h3")

for h3 in all_h3:
    print(h3)
```

```python
# to get the information
for h3 in all_h3:
    print(h3.get_text(strip=True))
```

```python
# find the parent of the tag
for h3 in all_h3:
    print(h3.find_parent("div")["class"])
```

```python
# to parse the next element to previous tag that is under same parent
for h3 in all_h3:
    print(h3.find_next("div").select("span.country-population")[0].get_text())
```

```python
all_countries = []
for h3 in all_h3:
    name = h3.get_text(strip=True)
    population = h3.find_next("div").select("span.country-population")[0].get_text()
    all_countries.append([name,population])
```

```python
all_countries
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(all_countries,columns = ["Country","Population"])
```
